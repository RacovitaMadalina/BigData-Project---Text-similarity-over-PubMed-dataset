```python
import gzip
import xml.etree.ElementTree as ET

input = gzip.open('../datasets/PubMed - medium samples/2016/2016-01-01_2016-01-02/records00000.xml.gz', 'r')
tree = ET.parse(input)
root = tree.getroot()

print(root.tag)
print(root.attrib)
```

```python
print(root.findall("."))
```

```python
for child in root:
    print(child.tag, child.attrib)
```

```python
set([elem.tag for elem in root.iter()])
```

```python
sorted([tag.split('}')[1] for tag in set([elem.tag for elem in root.iter()])])
```
